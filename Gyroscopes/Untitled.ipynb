```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0.5, 0.4, 0.3, 0.2, 0.1])
ypoints = np.array([35.5, 27.5, 23.1, 14.0, 9.50])

c = np.polyfit(xpoints, ypoints,1)
p = np.poly1d(c)

xp = np.linspace(min(xpoints), max(xpoints), 100)

plt.title('Time to fall squared vs height')
plt.ylabel(r"$t_F^2 \: (s^2)$")
plt.xlabel("Height (m)")
plt.plot(xpoints,ypoints, marker='o', linestyle='',color= 'c')
plt.plot(xp, p(xp), linestyle='--',color= 'c', label=f'y={round(c[0],2)}x+{round(c[1],2)}')
plt.legend(loc="upper left")
plt.savefig("plot1.pgf", bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0.03, 0.05, 0.07, 0.1, 0.12, 0.15])
yexppoints = np.array([0.822, 1.255, 1.753, 2.094, 2.887, 3.134])
ycalcpoints = np.array([0.173, 0.285, 0.395, 0.568, 0.701, 0.789])

plt.title('Angular velocity of precession vs mass')
plt.ylabel(r"$\omega_p \: (rads^{-1})$")
plt.xlabel(r"Mass (kg)")

c1 = np.polyfit(xpoints, yexppoints,1)
p1 = np.poly1d(c1)
c2 = np.polyfit(xpoints, ycalcpoints,1)
p2 = np.poly1d(c2)
xp = np.linspace(min(xpoints), max(xpoints), 100)

plt.plot(xpoints,yexppoints,linestyle= '',color= 'c',marker='o', label='Experimental')
plt.plot(xpoints,ycalcpoints, marker='o',linestyle='',color='goldenrod', label='Theoretical')
plt.plot(xp, p1(xp), linestyle='--',color='c')
plt.plot(xp, p2(xp), linestyle='--',color='goldenrod')
plt.legend(loc="upper left")
plt.savefig("plot2.pgf", bbox_inches='tight', pad_inches=0.1)
plt.show()
print(c1,c2)
```
