```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0.5, 0.4, 0.3, 0.2, 0.1])
ypoints = np.array([35.5, 27.5, 23.1, 14.0, 9.50])

c = np.polyfit(xpoints, ypoints,1)
p = np.poly1d(c)

xp = np.linspace(min(xpoints), max(xpoints), 100)

plt.title('Time to fall squared by height')
plt.ylabel(r"$t_F^2 \: (s^2)$")
plt.xlabel("Height (m)")
plt.plot(xpoints,ypoints, 'o')
plt.plot(xp, p(xp), 'b--', label=f'y={round(c[0],2)}x+{round(c[1],2)}')
plt.legend(loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([71.94, 76.03, 81.05, 84.19, 93.62])
yexppoints = np.array([0.1818, 0.1814, 0.1646, 0.1635, 0.1685])
ycalcpoints = np.array([0.0800, 0.0757, 0.0710, 0.0684, 0.0615])

plt.title('Angular Velocity of Spin vs Angular Velocity of Precession')
plt.ylabel(r"$\omega_s \: (rads^{-1})$")
plt.ylim(0.04,0.24)
plt.xlabel(r"$\omega_p \: (rads^{-1})$")

c1 = np.polyfit(xpoints, yexppoints,1)
p1 = np.poly1d(c1)
c2 = np.polyfit(xpoints, ycalcpoints,1)
p2 = np.poly1d(c2)
xp = np.linspace(min(xpoints), max(xpoints), 100)

plt.plot(xpoints,yexppoints, 'o', label='Experimental')
plt.plot(xpoints,ycalcpoints, 'o', label='Theoretical')
plt.plot(xp, p1(xp), '--')
plt.plot(xp, p2(xp), '--')
plt.legend(loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([82.94, 54.77])
ypoints = np.array([12.36, 8.59])

c = np.polyfit(xpoints, ypoints,1)
p = np.poly1d(c)

xp = np.linspace(min(xpoints), max(xpoints), 100)

plt.title('Angular velocity of nutation vs angular velocity of spin')
plt.ylabel(r"$\omega_n \: (rads^{-1})$")
plt.xlabel(r"$\omega_s \: (rads^{-1})$")
plt.plot(xpoints,ypoints, 'o')
plt.plot(xp, p(xp), 'b--')
plt.show()
```
