```python
import numpy as np
import matplotlib.pyplot as plt
```

## Part B 
(i)

```python
hbar = 1.055e-34
q = 1.602e-19
m = 9.109e-31 * 0.2
omega = 2e15
```

```python
# Generating my mesh
Np = 100
a = 1e-10 # m
X = a*np.linspace(-Np//2,Np//2,Np) #nm

# Define potential
V = 0.5*m*omega*omega*(X**2)

# Define Hamiltonian as a tridiagonal matrix
t0=(hbar*hbar)/(2*m*a*a)/q #divide by q to convert to eV
on=2.0*t0*np.ones(Np)
off=-t0*np.ones(Np-1)

H = np.diag(on+V) + np.diag(off,1) + np.diag(off,-1)
```

```python
# Solving for eigenvalues and eigenvectors

eigval,eigvec=np.linalg.eig(H)
idx = eigval.argsort()[::1]   
eigval = eigval[idx]
eigvec = eigvec[:,idx]
Energy_gap=eigval[1]-eigval[0]
print(Energy_gap)
Psi0=np.multiply(eigvec[:,0],eigvec[:,0])
Psi1=np.multiply(eigvec[:,1],eigvec[:,1])
```

```python
plt.plot(
    X,
    V,
    c='royalblue'
)

plt.title('Potential energy V(x)')
plt.xlabel('Distance (nm)')
plt.ylabel('Potential Energy (eV)')
plt.show()
```

```python
plt.plot(X,Psi0,c='royalblue')
plt.xlabel('Distance (nm)')
plt.ylabel('Probability')
plt.title(r'$|\Psi_0|^2$')
plt.show()
```

```python
plt.plot(X, Psi1,c='royalblue')
plt.xlabel('Distance (nm)')
plt.ylabel('Probability')
plt.title(r'$|\Psi_1|^2$')
plt.show()
```

```python
# Plot of difference between two eigenvalues as a function of alpha

#Changing alpha

omega = np.linspace(1e15,10e15,Np)

alpha = 0.5 * m * omega**2

energy_diff = []

for i in alpha:
    V = i * (X**2)
    t0=(hbar*hbar)/(2*m*a*a)/q #divide by q to convert to eV
    on=2.0*t0*np.ones(Np)
    off=-t0*np.ones(Np-1)

    H = np.diag(on+V) + np.diag(off,1) + np.diag(off,-1)
    eigval,eigvec=np.linalg.eig(H)
    idx = eigval.argsort()[::1]   
    eigval = eigval[idx]
    eigvec = eigvec[:,idx]
    energy_diff.append(eigval[1]-eigval[0])

# Theoretical function

def dE(alpha):
    return hbar*np.sqrt(2*alpha/m)/q

plt.plot(alpha, dE(alpha),c='orange',label='Expected')
plt.plot(alpha, energy_diff,c='royalblue',label='FDM')

plt.xlabel(r'$\alpha$')
plt.ylabel('Energy gap (eV)')
plt.title('Energy difference vs α')
plt.legend(loc='upper left')
plt.show()
```

```python
# Define new potential

r = 7 / 2 * 1e-9

alpha = (0.4e-3*q)**2*m/(2*hbar**2)

# No detuning bias

V1 = alpha * (X + r)**2
V2 = alpha * (X - r)**2
V_static = np.minimum(V1, V2)

t0=(hbar*hbar)/(2*m*a*a)
on=2.0*t0*np.ones(Np)
off=-t0*np.ones(Np-1)

H_static = np.diag(on+V_static) + np.diag(off,1) + np.diag(off,-1)

eigval,eigvec=np.linalg.eig(H_static)
idx = eigval.argsort()[::1]   
eigval = eigval[idx]
eigvec = eigvec[:,idx]
psi_0 = eigvec[:,0].copy()
psi_1 = eigvec[:,1].copy()

omega = (eigval[1] - eigval[0])/hbar

# With detuning
F = 1e11

t_min = 0
t_max = 2 * np.pi / (omega)
dt = t_max/500
t_range = np.arange(t_min,t_max,dt)

proj_0_t = []
proj_1_t = []

psi = psi_0.copy()

def H_build(potential):
    return np.diag(on+potential) + np.diag(off,1) + np.diag(off,-1)

for t in t_range:
    Vt = np.minimum(V1, V2) + F*X*q*np.cos(omega * t)
    Ht = H_build(Vt)

    c0 = np.vdot(psi_0, psi)
    c1 = np.vdot(psi_1, psi)

    p0 = np.abs(c0)**2
    p1 = np.abs(c1)**2

    proj_0_t.append(p0)
    proj_1_t.append(p1)

    Vt_next = np.minimum(V1, V2) + F*X*q*np.cos(omega * (t+dt))
    Ht_next = H_build(Vt_next)

    A = np.eye(Np) + 1j * dt/(2*hbar) * Ht_next
    B = np.eye(Np) - 1j * dt/(2*hbar) * Ht

    psi_new = np.linalg.solve(A,B @ psi)
    psi = psi_new / np.linalg.norm(psi_new)

proj_0_t = np.array(proj_0_t)
proj_1_t = np.array(proj_1_t)

plt.figure()
plt.plot(t_range, proj_0_t, label="Projection onto |0>",c='royalblue')
plt.plot(t_range, proj_1_t, label="Projection onto |1>",c='orange')
plt.xlabel("Time (s)")
plt.ylabel("Probability")
plt.xlim(0,3.5e-13)
plt.title("Projections of Wavefunction onto |0> and |1>")
plt.show()
```
