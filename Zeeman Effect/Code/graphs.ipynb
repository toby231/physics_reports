```python
# 1A

xpara = nopolpara1_0A['Radius']

ypara = nopolpara1_0A['Intensity']

peaks, _ = find_peaks(ypara,prominence=2,height=100)
for i in range(len(peaks)):
    plt.plot(peaks[i],ypara[peaks[i]],marker='x',color='red')

plt.scatter(xpara,ypara,marker='.',c='royalblue',label='Parallel')

plt.xlabel('Radius (px)')
plt.ylabel('Intensity')
plt.legend()
```

```python
# 1.5A

xpara = nopolpara1_5A['Radius']
xperp = nopolperp1_5A['Radius']

ypara = nopolpara1_5A['Intensity']
yperp = nopolperp1_5A['Intensity']

peaks, _ = find_peaks(ypara,prominence=2,height=100)
for i in range(len(peaks)):
    print(peaks[i],ypara[peaks[i]])

plt.scatter(xpara,ypara,marker='.',c='royalblue',label='Parallel')
plt.scatter(xperp,yperp,marker='.',c='orange',label='Perpendicular')

plt.xlabel('Radius (px)')
plt.ylabel('Intensity')
plt.legend()
```
